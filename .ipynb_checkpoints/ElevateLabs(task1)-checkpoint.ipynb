```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
titanic = pd.read_csv('Titanic Dataset.csv') #loading the datasets
```

```python
# Step 2: Initial data exploration
print("="*50)
print("Initial Data Exploration")
print("="*50)
print(f"Dataset shape: {titanic.shape}")
print("\nData types:")
print(titanic.dtypes)
print("\nMissing values:")
print(titanic.isnull().sum())
```

```python
# Visualize missing values
plt.figure(figsize=(10, 6))
sns.heatmap(titanic.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Step 3: Handle missing values
print("\n" + "="*50)
print("Handling Missing Values")
print("="*50)
```

```python
# Age: Fill with median grouped by Pclass and Sex
titanic['Age'] = titanic.groupby(['Pclass', 'Sex'])['Age'].transform(
    lambda x: x.fillna(x.median()))
```

```python
# Embarked: Fill with mode
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])
```

```python
# Cabin: Create new feature indicating whether cabin was known
titanic['Has_Cabin'] = titanic['Cabin'].notnull().astype(int)
titanic.drop('Cabin', axis=1, inplace=True)
```

```python
# Fare: Fill single missing value with median of class
titanic['Fare'] = titanic.groupby('Pclass')['Fare'].transform(
    lambda x: x.fillna(x.median()))
```

```python
# Verify missing values handled
print("\nMissing values after treatment:")
print(titanic.isnull().sum().sum())  # Should be 0
```

```python
# Step 4: Feature engineering
print("\n" + "="*50)
print("Feature Engineering")
print("="*50)
```

```python
# Create family size feature
titanic['Family_Size'] = titanic['SibSp'] + titanic['Parch'] + 1
titanic['Is_Alone'] = (titanic['Family_Size'] == 1).astype(int)
```

```python
# Extract titles from names
titanic['Title'] = titanic['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
titanic['Title'] = titanic['Title'].replace([
    'Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 
    'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
titanic['Title'] = titanic['Title'].replace('Mlle', 'Miss')
titanic['Title'] = titanic['Title'].replace('Ms', 'Miss')
titanic['Title'] = titanic['Title'].replace('Mme', 'Mrs')
```

```python
# Step 5: Convert categorical features
print("\n" + "="*50)
print("Encoding Categorical Features")
print("="*50)
```

```python
# One-hot encode categorical features
cat_features = ['Sex', 'Embarked', 'Title']
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_features = encoder.fit_transform(titanic[cat_features])
encoded_df = pd.DataFrame(
    encoded_features,
    columns=encoder.get_feature_names_out(cat_features)
)
```

```python
# Drop original columns and concatenate encoded features
titanic_processed = pd.concat([
    titanic.drop(cat_features, axis=1),
    encoded_df
], axis=1)
```

```python
# Drop unnecessary columns
titanic_processed.drop([
    'PassengerId', 'Name', 'Ticket', 
    'SibSp', 'Parch'
], axis=1, inplace=True)
```

```python
# Step 6: Handle numerical features
print("\n" + "="*50)
print("Scaling Numerical Features")
print("="*50)
```

```python
# Select numerical features to scale
num_features = ['Age', 'Fare', 'Family_Size']
scaler = StandardScaler()
titanic_processed[num_features] = scaler.fit_transform(
    titanic_processed[num_features])
```

```python
# Step 7: Outlier detection and treatment
print("\n" + "="*50)
print("Outlier Detection")
print("="*50)
```

```python
# Visualize outliers using boxplots
plt.figure(figsize=(15, 6))
for i, feature in enumerate(num_features, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=titanic_processed[feature])
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
```

```python
# Calculate IQR for Fare
Q1 = titanic_processed['Fare'].quantile(0.25)
Q3 = titanic_processed['Fare'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Cap outliers using IQR method
fare_upper_bound = Q3 + 1.5 * IQR
titanic_processed['Fare'] = np.where(
    titanic_processed['Fare'] > fare_upper_bound,
    fare_upper_bound,
    titanic_processed['Fare']
)
```

```python
# Step 8: Final dataset preview
print("\n" + "="*50)
print("Final Processed Dataset")
print("="*50)
print(f"Final shape: {titanic_processed.shape}")
print("\nFirst 5 rows:")
print(titanic_processed.head())
print("\nData types:")
print(titanic_processed.dtypes)
```
